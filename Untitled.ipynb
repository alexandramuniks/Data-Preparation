```python
missing_value_formats = ["n.a.","?","NA","n/a", "na", "--"]
df = pd.read_csv("spambase.csv", na_values = missing_value_formats)
```

```python
df['is_spam'] = df['is_spam'].map(make_int)
df['is_spam'].head(10)
```

```python
# Dropping irrelevant columns 
df = df.drop([‘Engine Fuel Type’, ‘Market Category’, ‘Vehicle Style’, ‘Popularity’, ‘Number of 
Doors’, ‘Vehicle Size’], axis=1) 
df.head(5)
```

```python
# Renaming the column names 
df = df.rename(columns={“Engine HP”: “HP”, “Engine Cylinders”: “Cylinders”, “Transmission 
Type”: “Transmission”, “Driven_Wheels”: “Drive Mode”,”highway MPG”: “MPG-H”, “city 
mpg”: “MPG-C”, “MSRP”: “Price” }) 
df.head(5) 
```

```python
# Rows containing duplicate data 
duplicate_rows_df = df[df.duplicated()] 
print(“number of duplicate rows: “, duplicate_rows_df.shape) 
```

```python
# Used to count the number of rows before removing the data -- usado para contar rows
df.count()  
```

```python
# Dropping the duplicates  
df = df.drop_duplicates() 
df.head(5)

# Finding the null values. 
print(df.isnull().sum()) 

# Dropping the missing values. 
df = df.dropna()  
df.count() 

# After dropping the values 
print(df.isnull().sum())  

*apos aplicar eda, explicar o pq foi aplicado. 
```

```python
# Plotting a Histogram 
df.Make.value_counts().nlargest(40).plot(kind=’bar’, figsize=(10,5)) 
plt.title(“Number of cars by make”) 
plt.ylabel(‘Number of cars’) 
plt.xlabel(‘Make’);
```

```python
# Plotting a scatter plot - mapa de dispersão
fig, ax = plt.subplots(figsize=(10,6)) 
ax.scatter(df[‘HP’], df[‘Price’]) 
ax.set_xlabel(‘HP’) 
ax.set_ylabel(‘Price’) 
plt.show() 
```
